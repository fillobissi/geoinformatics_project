#### Prova salvataggio automatico dei dati

```python
# Inizializza il client DDSAPI
c = ddsapi.Client(key="1215a53b-ffb6-485b-8aa5-ae67eba56758:VxHqCSyC5X1m9Y3y2ktoXpq1TPO_JnYpLWcsyWa1w5A")  # Inserisci la tua API key qui

# Effettua una richiesta per scaricare i dati
request = c.retrieve(
    "era5-downscaled-over-italy", "hourly",
    {
        "area": {
            "north": 45.5,
            "south": 45.4,
            "east": 9.3,
            "west": 9
        },
        "time": {
            "year": ["2023"],
            "month": ["8"],
            "day": [str(i) for i in range(16, 17)],
            "hour": [str(i).zfill(2) for i in range(0, 24)]
        },
        "variable": ["air_temperature"],
        "format": "netcdf"
    },
    "era5-downscaled-over-italy-hourly.nc"
)

# Controlla ripetutamente se la richiesta è pronta
while request.status != "done":
    print("I dati non sono ancora pronti.")
    time.sleep(300)  # Attendi 300 secondi prima di verificare di nuovo lo stato
    request.update()  # Aggiorna lo stato della richiesta

# Controlla se la richiesta è pronta
if request.status == "done":
    download_url = request.location
    
    # Aggiungi timestamp al nome del file
    timestamp = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
    output_file = rf"C:\Users\andre\OneDrive - Politecnico di Milano\Geoinformatics_project\data\data_downloads\initial_temps\salvataggio_automatico\era5-downscaled-{timestamp}.nc"
    
    # Scarica il file a pezzi di 1kb per alleggerire
    response = requests.get(download_url, stream=True)
    with open(output_file, 'wb') as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
    print(f"Download completato: {output_file}")
```
