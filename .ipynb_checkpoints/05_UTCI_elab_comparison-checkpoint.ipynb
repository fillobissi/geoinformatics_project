### 05 - UTCI ELAB COMPARISON

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import random
from pythermalcomfort.models import utci as utci_full

# ===============================
# LOAD DATASETS
# ===============================
# dataset1 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22197.nc")
# dataset2 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22209.nc")
# dataset3 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22223.nc")
# dataset4 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22224.nc")
# dataset5 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22228.nc")
# dataset6 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22230.nc")

ds_u10 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22197.nc")   # U_10M
ds_td = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22209.nc")    # TD_2M + ATHB_S
ds_ta = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22223.nc")    # T_2M
ds_rad_sw = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22228.nc")  # ASOB_S
ds_v10 = xr.open_dataset(r"C:\Users\andre\OneDrive\Desktop\Datasets_Geoinfprj\era5-downscaled-over-italy_hourly_22230.nc")   # V_10M

# ===============================
# SELECT RANDOM TIMESTAMP
# ===============================
timestamps = ds_ta['T_2M'].time.values
random_timestamp = random.choice(timestamps)

# ===============================
# EXTRACT VARIABLES
# ===============================
Ta_k = ds_ta['T_2M'].sel(time=random_timestamp)
Td_k = ds_td['TD_2M'].sel(time=random_timestamp)
U_10M = ds_u10['U_10M'].sel(time=random_timestamp)
V_10M = ds_v10['V_10M'].sel(time=random_timestamp)
LW = ds_td['ATHB_S'].sel(time=random_timestamp)
SW = ds_rad_sw['ASOB_S'].sel(time=random_timestamp)

# Convert Kelvin to Celsius
Ta_c = Ta_k - 273.15
Td_c = Td_k - 273.15

# Relative Humidity
def calculate_relative_humidity(Ta_c, Td_c):
    RH = 100.0 * np.exp((17.625 * Td_c) / (243.04 + Td_c) - (17.625 * Ta_c) / (243.04 + Ta_c))
    return RH

RH = calculate_relative_humidity(Ta_c, Td_c)

# Wind speed
V_wind = np.sqrt(U_10M**2 + V_10M**2)

# Mean radiant temperature from longwave + shortwave radiation
sigma = 5.67e-8  # Stefan-Boltzmann constant
MRT_K = ((LW + SW) / sigma) ** 0.25
MRT_C = MRT_K - 273.15

# ===============================
# SIMPLIFIED UTCI
# ===============================
def calculate_utci_simple(Ta_c, RH):
    return Ta_c + 0.607562 + 0.022771 * Ta_c - 0.003578 * RH - 0.000119 * Ta_c * RH

utci_simple = calculate_utci_simple(Ta_c, RH)

# ===============================
# OFFICIAL UTCI
# ===============================
Ta_flat = Ta_c.values.flatten()
tr_flat = MRT_C.values.flatten()
v_flat = V_wind.values.flatten()
rh_flat = RH.values.flatten()

utci_official_flat = utci_full(tdb=Ta_flat, tr=tr_flat, v=v_flat, rh=rh_flat)
utci_official = utci_official_flat.reshape(Ta_c.shape)

# ===============================
# PLOTTING
# ===============================
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
im1 = axs[0].imshow(utci_simple, cmap='coolwarm')
axs[0].set_title('Simplified UTCI')
plt.colorbar(im1, ax=axs[0])

im2 = axs[1].imshow(utci_official, cmap='coolwarm')
axs[1].set_title('Official UTCI (pythermalcomfort)')
plt.colorbar(im2, ax=axs[1])

plt.suptitle(f'UTCI Comparison at {np.datetime_as_string(random_timestamp, unit="h")}')
plt.tight_layout()
plt.show()

# ===============================
# STATISTICS
# ===============================
def print_statistics(data, label):
    print(f"\nStatistics for {label}:")
    print(f"Mean: {np.nanmean(data):.2f} °C")
    print(f"Median: {np.nanmedian(data):.2f} °C")
    print(f"95th Percentile: {np.nanpercentile(data, 95):.2f} °C")
    print(f"99th Percentile: {np.nanpercentile(data, 99):.2f} °C")
    print(f"Max: {np.nanmax(data):.2f} °C")
    print(f"Min: {np.nanmin(data):.2f} °C")

print_statistics(utci_simple, "Simplified UTCI")
print_statistics(utci_official, "Official UTCI (pythermalcomfort)")
```
